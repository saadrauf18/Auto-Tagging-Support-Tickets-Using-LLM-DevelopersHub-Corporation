```python
from transformers import pipeline
import gradio as gr
```

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
possible_labels = ["Billing Issue", "Technical Problem", "Account Management",
                   "Service Outage", "General Inquiry", "Feature Request"]
```

```python
sample_ticket = "My internet connection keeps dropping every few minutes since last night."
result = classifier(sample_ticket, candidate_labels=possible_labels)
print(result)
```

```python
def auto_tag(ticket_text):
    result = classifier(ticket_text, candidate_labels=possible_labels)
    top_label = result["labels"][0]
    top_score = result["scores"][0]
    return f"Predicted Tag: {top_label} (Confidence: {top_score:.2f})"
```

```python
demo = gr.Interface(
    fn=auto_tag,
    inputs=gr.Textbox(lines=3, placeholder="Enter support ticket text..."),
    outputs="text",
    title="Auto-Tagging Support Tickets (Zero-Shot LLM)",
)
demo.launch(share=True)
```
